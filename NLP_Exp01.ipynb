```python
#Sentence Tokenization
```

```python
from nltk.tokenize import sent_tokenize
```

```python
text = '''In the heart of the bustling city, amidst the chaotic rhythm of daily life, there exists a serene oasis tucked away like a well-kept secret. A small, quaint café with its weathered wooden sign beckons passersby with the promise of aromatic coffee and freshly baked pastries. Inside, the air is alive with the gentle hum of conversation and the soft clinking of porcelain against saucers. Sunlight filters through lace curtains, casting intricate patterns on the worn, yet inviting, tabletops. Here, time seems to slow down, allowing patrons to savor each sip and appreciate the simple pleasure of a quiet moment away from the city's relentless pace.'''
```

```python
text
```

```python
sentences = sent_tokenize (text)
```

```python
sentences
```

```python
#Word Tokenization
```

```python
from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize (text)
```

```python
words
```

```python
for w in words:
    print (w)
```

```python
#Levels of Sentences Tokenization using Comprehension
```

```python
sent_tokenize (text)
```

```python
[word_tokenize (text) for t in sent_tokenize(text)]
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize (text)
```
